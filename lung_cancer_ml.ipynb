#CLUSTERING

```python
from sklearn.cluster import KMeans
cluster = KMeans(n_clusters=3,random_state=42,n_init=10)
features = [
    'Age',
    'Air Pollution',
    'Alcohol use',
    'Dust Allergy',
    'OccuPational Hazards',
    'Genetic Risk',
    'Smoking',
    'Passive Smoker',
    'chronic Lung Disease',
    'Chest Pain',
    'Coughing of Blood',
    'Fatigue',
    'Weight Loss',
    'Shortness of Breath',
    'Wheezing',
    'Swallowing Difficulty',
    'Clubbing of Finger Nails',
    'Frequent Cold',
    'Dry Cough'
]
X = df[features]
df['level_clusstered'] = cluster.fit_predict(X)
```

```python
df['level_clusstered']
```

```python
df[['Level','level_clusstered']].drop_duplicates()
```

```python
df.columns
```

# HERE WE CAN SEE THAT AFTER CLUSTERING ALL THE LEVELS ARE CLUSSTERED IN ALL THE THREE CLUSSTERS THAT WHY SOMETIMES CLUSSTERING FAILS 

```python
plt.figure()
plt.scatter(df['Level'],df['level_clusstered'])
```

# NOW LETS MAKE LUNG CANCER PREDICTION MODEL BY THE LOGISTIC REGRESSION

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
label = LabelEncoder()
model = LogisticRegression()
df['level_encoded'] = label.fit_transform(df['Level'])
df['level_encoded']
```

```python
df[['Level','level_encoded']].drop_duplicates().sort_values('level_encoded')
```

```python
input_features = [ 
    'Age',
    'Gender',
    'Air Pollution',
    'Alcohol use',
    'Dust Allergy',
    'OccuPational Hazards',
    'Genetic Risk',
    'Smoking',
    'Passive Smoker',
    'chronic Lung Disease',
    'Chest Pain',
    'Coughing of Blood',
    'Fatigue',
    'Weight Loss',
    'Shortness of Breath',
    'Wheezing',
    'Swallowing Difficulty',
    'Clubbing of Finger Nails',
    'Frequent Cold',
    'Dry Cough'
]
y = df['level_clusstered']
Z = df[input_features]
model.fit(Z,y)
```

```python
age = int(input('enter your age: '))
gender = int(input('enter your gender: '))
air_pollution = int(input('enter air pollution level of your city(1 to 8):'))
alcohol_use = int(input('enter use level per day (1 to 8):'))
dust_allergy = int(input('dust allergy level (1 to 8):'))
occupational_hazzard = int(input('occupational hazzards level (1 to 8 ):'))
genetic_risk = int(input('genetic risk of lung disease level (1 to 7):'))
smoking = int(input('smoking level (1 to 8):'))
passive_smoker = int(input('passive smoker level (1 to 8 ):'))
chronic_lung_disease = int(input('chronic lung disease level (1 to 7):'))
chest_pain = int(input('chest pai type (1 to 9):'))
coughing_blood = int(input('coughing of blood level (1 to 9):'))
fatigue = int(input('fatigue level (1 to 9):'))
weight_loss = int(input('weight loss level (1 to 8):'))
shortness_of_breath = int(input('shortness of breath level (1 to 9):'))
wheezing = int(input('wheezing level (1 to 8):'))
swallowing_difficulty = int(input('swallowing difficulty level (1 to 8):'))
clubbing_of_finger_nails = int(input('clubbing of finger nails (1 to 9):'))
frequent_cold = int(input('frequent cold level (1 to 7):'))
dry_cough = int(input('dry cough level (1 to 7):'))
lvl = model.predict([[age,gender,air_pollution,alcohol_use,dust_allergy,occupational_hazzard,genetic_risk,smoking,passive_smoker,chronic_lung_disease,chest_pain,coughing_blood,fatigue,weight_loss,shortness_of_breath,wheezing,swallowing_difficulty,clubbing_of_finger_nails,frequent_cold,dry_cough]])
lvl = label.inverse_transform(lvl)[0]
print(f'After doing this analysis the machine predicted that the lung cancer level of the following patient is {lvl}')
print(f''' 
    'Age' {age},
    'Gender' {gender},
    'Air Pollution' {air_pollution},
    'Alcohol use' {alcohol_use},
    'Dust Allergy' {dust_allergy},
    'OccuPational Hazards' {occupational_hazzard},
    'Genetic Risk' {genetic_risk},
    'Smoking' {smoking},
    'Passive Smoker' {passive_smoker},
    'chronic Lung Disease' {chronic_lung_disease},
    'Chest Pain' {chest_pain},
    'Coughing of Blood' {coughing_blood},
    'Fatigue' {fatigue},
    'Weight Loss' {weight_loss},
    'Shortness of Breath' {shortness_of_breath},
    'Wheezing' {wheezing},
    'Swallowing Difficulty {swallowing_difficulty}',
    'Clubbing of Finger Nails' {clubbing_of_finger_nails},
    'Frequent Cold' {frequent_cold},
    'Dry Cough' {dry_cough}
''')
```

#ACCURACY SCORE

```python
from sklearn.metrics import f1_score,recall_score
y_pred = model.predict(df[input_features])
```

```python
y_true = df['level_encoded']
```

# HENCE THE NO OF FEATURES ARE LARGE , SO WE USE F1 SCORE AND RECALL SCORE RATHER THAN ACCURACY SCORE

```python
f1 = f1_score(y_true,y_pred,average='weighted')
recall = recall_score(y_true,y_pred,average='weighted')
print(f'Therefore the model f1 score is {f1} and recall score is {recall}')
```
